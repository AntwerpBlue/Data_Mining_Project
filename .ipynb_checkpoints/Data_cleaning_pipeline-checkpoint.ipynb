```python
import pandas as pd
import numpy as np

df = pd.read_csv("Dataset/bank_marketing_train.csv")
print("df.shape = " ,df.shape)
df.head()
```

```python
df.info()
```

```python
# 避免污染原始数据，先复制一个数据
df_clean = df.copy()
```

```python
df_clean.columns.tolist()
```

```python
# 把999改为-1，方便后面标准化
df_clean['pdays'] = df_clean['pdays'].replace(999, -1)
df_clean['pdays'].value_counts().head()
```

```python
# 观察类别不平衡情况
df_clean['y'].value_counts(normalize=True)
```

```python
unknown_counts = (df_clean == "unknown").sum()
unknown_counts[unknown_counts > 0]
```

```python
# 把y变成0/1变量
df_clean['y'] = df_clean['y'].map({'yes': 1, 'no': 0})
df_clean['y'].value_counts()
```

```python
# 找出数值列
numeric_cols = df_clean.select_dtypes(include=['int64','float64']).columns  #include包括，exclude删除
numeric_cols
```

```python
# 查看每个数值列的大致统计特征
df_clean[numeric_cols].describe()
```

```python
# 发现F1有异常值，处理掉
df_clean['feature_1'].replace([np.inf, -np.inf], np.nan).isna().sum()
```

```python
# 中位数替换异常值
df_clean['feature_1'] = df_clean['feature_1'].replace([np.inf, -np.inf], np.nan)
median_val = df_clean['feature_1'].median()
df_clean['feature_1'] = df_clean['feature_1'].fillna(median_val)
```

```python
# 检查是否还有异常值
np.isinf(df_clean.select_dtypes(include=['float64', 'int64'])).sum()
```

```python
# 检查是否还有缺失值
df_clean.isna().sum()
```

```python
categorical_cols = df_clean.select_dtypes(include=['object']).columns.tolist()
Labelcols = set(df_clean.columns.tolist())-set(categorical_cols)
Labelcols = list(Labelcols)
Labelcols
```

```python
education_map = {
    'unknown': 0, 'illiterate': 1, 'basic.4y': 2, 'basic.6y': 3,
    'basic.9y': 4, 'high.school': 5, 'professional.course': 6, 'university.degree': 7
}

month_map = {
    'jan': 1, 'feb': 2, 'mar': 3, 'apr': 4, 'may': 5, 'jun': 6,
    'jul': 7, 'aug': 8, 'sep': 9, 'oct': 10, 'nov': 11, 'dec': 12
}

day_of_week_map = {
    'mon': 1, 'tue': 2, 'wed': 3, 'thu': 4, 'fri': 5
}

df_clean['education'] = df_clean['education'].map(education_map)
df_clean['month'] = df_clean['month'].map(month_map)
df_clean['day_of_week'] = df_clean['day_of_week'].map(day_of_week_map)
```

```python
df_clean.head()
```

```python
Nominalcols = [
    'job', 'marital', 'default', 'housing', 'loan', 
    'contact', 'poutcome', 'feature_3', 'feature_4', 'feature_5'
]
df_clean_processed = pd.get_dummies(df_clean, columns=Nominalcols, drop_first=False)
df_clean_processed.head()
```

```python
i = 0
for col in df_clean_processed.columns:
    print(col, end = '\t')
    i += 1
    if i % 5 == 0:
        print()
```

```python
df_lrsvm_ready = df_clean.copy()
categorical_cols
```

```python
df_lrsvm_ready = pd.get_dummies(
    df_lrsvm_ready,
    columns=categorical_cols,
    drop_first=False
)
df_lrsvm_ready.info()
```

```python
# 将所有 bool 类型的列转为 float，提高模型兼容性
bool_cols = df_lrsvm_ready.select_dtypes(include=['bool']).columns
df_lrsvm_ready[bool_cols] = df_lrsvm_ready[bool_cols].astype(float)

df_lrsvm_ready.info()
```

```python
# 1. y 是标签列
y_lrsvm = df_lrsvm_ready['y']

# 2. X 是除了 y 之外的所有特征
X_lrsvm = df_lrsvm_ready.drop(columns=['y'])

# 查看一下结果
X_lrsvm.shape, y_lrsvm.shape
```

```python
from sklearn.preprocessing import StandardScaler

# 1. 创建 StandardScaler 实例
#    参数说明：
#    - with_mean=True：对每列减去均值（默认）
#    - with_std=True：除以标准差（默认）
#    这是最常用的 Z-score 标准化
scaler = StandardScaler()

# 2. 使用训练集拟合 scaler（学习每一列的 mean 和 std）
#    注意：这里不能对 test 数据 fit！否则数据泄露。
scaler.fit(X_lrsvm)

# 3. 对训练集进行标准化，把数值缩放到 N(0,1)
X_lrsvm_scaled = scaler.transform(X_lrsvm)

# 4. transform 输出是 numpy array，我们转回 DataFrame，保持列名一致
X_lrsvm_scaled = pd.DataFrame(
    X_lrsvm_scaled,
    columns=X_lrsvm.columns,
    index=X_lrsvm.index
)

# 查看形状确认
X_lrsvm_scaled.shape
```
